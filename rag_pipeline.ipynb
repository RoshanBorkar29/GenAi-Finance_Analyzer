```python
from pathlib import Path
from langchain_core.documents import Document
```

```python
processed_folder=Path("data/processed")
image_desc_folder=Path("data/extracted_text")
markdown_file=processed_folder/"NYSE_HRL_2025.md"
with open(markdown_file,"r",encoding="utf-8") as f:
    markdown_text=f.read()       

type(markdown_text) 
```

```python
print(markdown_text[:500])  # Print the first 500 characters of the markdown text
```

```python
documents=[]
documents.append(
    Document(
        page_content=markdown_text,
        metadata={
            "type": "markdown",
            "source":"NYSE_HRL_2025.md"
        }
    )
)
print(documents)
len(documents)
```

```python
for txt_file in image_desc_folder.glob("*.txt"):
    with open(txt_file,"r",encoding="utf-8") as f:
        image_description=f.read()
        documents.append(
            Document(
                page_content=image_description,
                metadata={
                    "type":"image",
                    "source":txt_file.name
                }
            )
        )
print(len(documents))
```

```python
print(documents[4])
```

```python
%pip install langchain-text-splitters
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)
```

```python
chunks=text_splitter.split_documents(documents)
```

```python
print(len(chunks))
```

```python
print(chunks[7])
```

```python
%pip install sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer
embedding_model=SentenceTransformer("all-MiniLM-L6-v2")

embedding=embedding_model.encode("Revenue increased by 18%")
print(type(embedding))
print(len(embedding))
```

```python
embeddings=[]
for chunk in chunks:
    vector=embedding_model.encode(chunk.page_content)
    embeddings.append(
        vector
    )
```

```python
print(embeddings[0].shape)
```

```python
vector_store=[]
for chunk in chunks:
    vector=embedding_model.encode(chunk.page_content)
    vector_store.append({
        "embedding": vector,
        "document":chunk
    })
```

```python
%pip install faiss-cpu
```

```python
import faiss
import numpy as np
```

```python
type(embeddings[0])
```

```python
print(embeddings[0].shape)
```

```python
embeddings=np.array(
    embeddings,
    dtype=np.float32
)
```

```python
print(embeddings.dtype)
```

```python
dimension=embeddings.shape[1]
index=faiss.IndexFlatL2(dimension)##flat stores all vectors exctly no compression most accurate
```

```python
index.add(embeddings)
print(index.ntotal)
```

```python
faiss.write_index(index,"vector_store/finance_report.index")
```

```python
document_store=chunks
print(len(document_store))
```

```python
query="Why did revenue increase in 2025?"
query_embedding=embedding_model.encode(query)
print(query_embedding.shape)
```

```python
##faiss expectrs (1,384) so we will reshape it
query_embedding=np.array([query_embedding],dtype=np.float32)
print(query_embedding.shape)
```

```python
k=5
distance,indices=index.search(query_embedding,k)
print(indices)
print(distance)
```
